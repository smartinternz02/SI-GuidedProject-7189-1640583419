```python
import numpy as np
import pandas as pd
```

```python
data_set= pd.read_csv("zomato.csv")
```

```python
print(data_set)
```

```python
data_review=data_set['reviews_list']
```

```python
print(data_set['reviews_list'][0])
```

```python
x=[]
y=[]
```

```python
for row_num in range(0, 51717):
    #splitting the review text at '()
    lst= data_review[row_num].split("('")
    for i in lst:
        if len(i) > 5:
            if i.find("',")!=-1:
                single_rev = i.split("',")
                if len(single_rev[0])>2:
                    x.append(single_rev[0])
                if len(single_rev[1])>2 :
                    y.append(single_rev[1])
    
```

```python
import re
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
# to store the final rating
final_rating=[]
# to store the final reviews
final_review=[]
```

```python
for j in range(0,40000):
    data_x=x[j]
    data_x=re.sub('[a-zA-Z]'," ",data_x)
    data_x = ''.join(data_x)
    data_x = float(data_x)
    if data_x <2.5:
        final_rating.append("poor")
    elif data_x>=2.5 and data_x<=3.5:
        final_rating.append("average")
    elif data_x > 3.5:
        final_rating.append("good")
```

```python
print(final_rating)
```

```python
# label encoding the ratings and one hot encoding the classification
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
final_rating= le.fit_transform(final_rating)
```

```python
final_rating=np.array(final_rating)
final_rating= final_rating.reshape(-1)
final_rating= np.expand_dims(final_rating,axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
one=OneHotEncoder()
rates= one.fit_transform(final_rating).toarray()
```

```python
#Removing the stopwords 
#Stemming also done
for loop in range(0,40000):
    data_y= y[loop]
    data_y=re.sub('[^a-zA-Z]'," ",data_y)
    data_y=data_y.lower()
    data_y = data_y.split()
    data_y= [ps.stem(word) for word in data_y if not word in set(stopwords.words('english'))]
    data_y=' '.join(data_y)
    final_review.append(data_y)
```

```python
#count vectorize the reviews according to unique words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features = 20000)
x_final= cv.fit_transform(final_review).toarray()
```

```python
#saving the vectorizer ehich would be used as dictionary
import pickle
pickle.dump(cv,open('cv.pkl','wb'))
```

```python
#Spliting the data into testing and training data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_final,rates,test_size=0.2,random_state=0)
```

```python
#adding the neuron layers
#the inputlayers is equal to the number of unique words
# we take 3 deep layers  of 2000 units each
# relu activation is used
#output layer has 3 units as one hot encoding has 3 columns
# the classification is categorical
```

```python
from keras.models import Sequential
from keras.layers import Dense
model=Sequential()
model.add(Dense(units=13264,kernel_initializer='random_uniform',activation='relu'))
model.add(Dense(units=2000, kernel_initializer='random_uniform',activation='relu'))
model.add(Dense(units=2000, kernel_initializer='random_uniform',activation='relu'))
model.add(Dense(units=2000, kernel_initializer='random_uniform',activation='relu'))
model.add(Dense(units=3, kernel_initializer='random_uniform',activation='softmax'))
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,batch_size=128,epochs=10)
```

```python
#testing the prediction
y_pred = model.predict(x_test)

text= "The food is okay average place"
text = re.sub('[^a-zA-Z]',' ',text)
text= text.lower()
text= text.split()
text = [ps.stem(word) for word in text if not word in set(stopwords.words('english'))]
text= ' '.join(text)
```
